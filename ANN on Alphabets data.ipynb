```python
import pandas as pd
data = pd.read_csv('Alphabets_data.csv')
data.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
import numpy as np
```

```python

# Separate features and target
X = data.drop(columns='letter')
y = data['letter']

# Normalize the feature data
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)
```

```python
import tensorflow as tf

from scikeras.wrappers import KerasClassifier, KerasRegressor
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split

# Divide the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)

# Convert target labels to one-hot encoding
y_train_encoded = tf.keras.utils.to_categorical(pd.factorize(y_train)[0])
y_test_encoded = tf.keras.utils.to_categorical(pd.factorize(y_test)[0])
```

```python
# Construct a basic ANN model
model = Sequential([
    Dense(64, input_shape=(X_train.shape[1],), activation='relu'),
    Dense(32, activation='relu'),
    Dense(y_train_encoded.shape[1], activation='softmax')  # Output layer
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train_encoded, epochs=20, validation_split=0.2, batch_size=32)
```

```python
# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(X_test, y_test_encoded)
print(f"Test Accuracy: {test_accuracy:.4f}")
# Predict on the test set
y_pred = model.predict(X_test)
y_pred_classes = y_pred.argmax(axis=1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.model_selection import GridSearchCV

# Custom KerasClassifier class
class KerasClassifier(BaseEstimator, ClassifierMixin):
    def __init__(self, build_fn=None, **kwargs):
        self.build_fn = build_fn
        self.params = kwargs
        self.model = None

    def fit(self, X, y, **kwargs):
        self.model = self.build_fn(**self.params)
        self.model.fit(X, y, **kwargs)
        return self

    def predict(self, X, **kwargs):
        return self.model.predict(X, **kwargs).argmax(axis=1)

    def predict_proba(self, X, **kwargs):
        return self.model.predict(X, **kwargs)

    def score(self, X, y, **kwargs):
        y_pred = self.predict(X, **kwargs)
        return np.mean(y_pred == y.argmax(axis=1))

from scikeras.wrappers import KerasClassifier

# Function to create the model
def create_model(optimizer='adam', neurons=32, activation='relu'):
    model = Sequential([
        Dense(neurons, input_dim=X_train.shape[1], activation=activation),
        Dense(neurons, activation=activation),
        Dense(y_train_encoded.shape[1], activation='softmax')
    ])
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Wrap the model using KerasClassifier from scikeras.wrappers
model = KerasClassifier(model=create_model, verbose=0)

# Define the grid of hyperparameters
param_grid = {
    'batch_size': [16, 32],
    'epochs': [10, 20],
    'model__optimizer': ['adam', 'sgd'],  # Notice the 'model__' prefix
    'model__neurons': [32, 64],          # Notice the 'model__' prefix
    'model__activation': ['relu', 'tanh'] # Notice the 'model__' prefix
}

# Perform Grid Search
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=3)
grid_result = grid.fit(X_train, y_train_encoded)

# Display the best hyperparameters
print(f"Best parameters found: {grid_result.best_params_}")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
# Get predicted class labels from probabilities
y_pred_classes = np.argmax(y_pred, axis=1)

# Convert predicted labels to strings to match y_test
y_pred_classes = y_pred_classes.astype(str)

# Now calculate the metrics
accuracy = accuracy_score(y_test, y_pred_classes)
precision = precision_score(y_test, y_pred_classes, average='macro')
recall = recall_score(y_test, y_pred_classes, average='macro')
f1 = f1_score(y_test, y_pred_classes, average='macro')

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```
